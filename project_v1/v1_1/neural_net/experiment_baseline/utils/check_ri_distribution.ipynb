This notebook creates a plot of the distributions of lower and upper RI limits after standardization. The purpose is to show that the RI's are distributions across a continuous range, instead of only at discrete points, which would not be suitable for an ML model to learn to generalize outside of the training data. 

```python
import numpy as np
import matplotlib.pyplot as plt

y = np.load('../data/y.npy')
y = y[:,:2] # only get RI limits

y.shape
```

```python
plt.figure(figsize=(5,3), dpi=100)
plt.hist(y[:,0], np.linspace(-3, 3, 100), alpha=0.5, label='Lower');
plt.hist(y[:,1], np.linspace(-3, 3, 100), alpha=0.5, label='Upper');
plt.gcf().set_facecolor('white')
plt.xlabel('Limit value')
plt.ylabel('Frequency')
plt.legend()
plt.xlim([-2, 3]);
```
