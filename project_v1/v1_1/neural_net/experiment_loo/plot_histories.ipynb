```python
import os
import pickle
import matplotlib.pyplot as plt

holdout_analytes = ['Hb', 'Ca', 'FT4', 'AST', 'LACT', 'GGT', 'TSH', 'IgE', 'CRP', 'LDH']

plt.figure(figsize=(10, 4))
for c, an in enumerate(holdout_analytes):
    
    with open('./models_nodrop/model_'+an+'/model_history.npy', 'rb') as file_pi:
        history = pickle.load(file_pi)
    plt.subplot(2,5,c+1)
    plt.plot(history['loss'], label='nodrop loss')
    plt.plot(history['val_loss'], label='nodrop val_loss')
    with open('./models_dropout/model_'+an+'/model_history.npy', 'rb') as file_pi:
        history = pickle.load(file_pi)
    plt.plot(history['loss'], label='drop loss')
    plt.plot(history['val_loss'], label='drop val_loss')
    if c==0:
        plt.legend()
        plt.ylabel('Mean Absolute Error')
    if c==5:
        plt.xlabel('Epoch')
    plt.title(an)
    plt.ylim([0, 0.5])
plt.gcf().set_facecolor('white');
plt.gcf().tight_layout();

```
